```python
import imp
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import  seaborn as sns
```

```python
cols = ["sepal_length", "sepal_width", "petallength", "petalwidth","class"]
```

```python
df = pd.read_csv("iris.data.csv", names=cols)
df.shape
```

```python
df["class"].value_counts()
```

```python
# histogram visualisation 
from matplotlib.pyplot import plot


df.hist(figsize=(12, 6))
plot()
```

```python
#boxplot using seaborn 
sns.boxplot(x="sepal_length", y="sepal_width", data=df)
plt.title("petal distribution of dataset")
plt.show()
```

```python
# scatterplot visualisation of the three species 
sns.scatterplot(x ='sepal_length',  y ='sepal_width', hue = "class", data = df)
plt.title('a scatterplot for the three different species'),
plt.show()
```

```python
# Barchart visualisation 
sns.countplot(x='class', data=df)
plt.title("Barchart")
plt.show()
```

```python
# corrolation visualisation 
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("A corrolation of iris dataset")
```

```python
# pie chart visualisation 
from attr import dataclass


plt.pie(df["class"].value_counts(), labels=df["class"].unique())
plt.title("piechart representation")
plt.show()
```
