```python
import pandas as pd 
import matplotlib.pyplot as plt
from  sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import seaborn as sns 
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
df = pd.read_csv("petrol_consumption (1).csv")
df.columns
```

```python
df.hist(figsize=(6, 12))
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap="Accent", linewidths=0.5)
plt.title("Corrolation Metrix Heatmap")
plt.show()
```

```python
# Splittng the dataset

X = df.drop(columns=["Petrol_Consumption"])
y = df["Petrol_Consumption"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape, X_test.shape, y_train,y_test
```

```python
# Normalization

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
# building the model 

ran_model = RandomForestRegressor()
ran_model = ran_model.fit(X_train, y_train)
y_pred = ran_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
Dtree_model = DecisionTreeRegressor()
Dtree_model = Dtree_model.fit(X_train, y_train)
y_pred = Dtree_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error


mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
Kn_model = KNeighborsRegressor()
Kn_model = Kn_model.fit(X_train, y_train)
y_pred = Kn_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```
